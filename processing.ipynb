### Data 2

```python
dbsh = list(range(1, 13))
td_pb = list(range(13, 28))
btb_dhmt = list(range(28,43))
tn = list(range(43,49))
dnb = list(range(49,56))
dbscl = list(range(56,70))
```

```python
plt.figure(figsize=(20, 10))
bar_width = 0.2

plt.subplot(1, 2, 1)
for i, year in enumerate(['2018', '2020', '2021']):
        plt.bar([p + bar_width * i for p in range(len(dbsh))], 
            data_2[f'Tổng diện tích - {year}'][dbsh], 
            width=bar_width, 
            label=f'Tổng diện tích - {year}')
        plt.title('Đồng bằng sông Hồng')
        plt.xticks([p + bar_width for p in range(0, len(dbsh))], data_2['Vị trí'][dbsh], rotation=90)

plt.subplot(1, 2, 2)
for i, year in enumerate(['2018', '2020', '2021']):
        plt.bar([p + bar_width * i for p in range(len(td_pb))], 
            data_2[f'Tổng diện tích - {year}'][td_pb], 
            width=bar_width, 
            label=f'Tổng diện tích - {year}')
        plt.title('Trung du và miền núi phía Bắc')
        plt.xticks([p + bar_width for p in range(0, len(td_pb))], data_2['Vị trí'][td_pb], rotation=90)

plt.xlabel('Vị trí')
plt.ylabel('Diện tích (ha)')
plt.suptitle('Tổng diện tích đất qua các năm')
plt.legend(loc='upper right', bbox_to_anchor=(1,1.15))
plt.tight_layout()
plt.show()
```
